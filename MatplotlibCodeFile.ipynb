```python
normals = pd.Series(np.random.normal(size=50))

normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
variables = pd.DataFrame({'normal': np.random.normal(size=30), 
                       'gamma': np.random.gamma(1, size=30), 
                       'poisson': np.random.poisson(size=30)})
print(variables)

variables.cumsum(0).plot()
```

```python
variables.cumsum(0).plot(subplots=True)
```

```python
variables.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(6, 12))
for i,var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

```python
rng = np.random.RandomState(7)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,
            cmap='Blues')
plt.colorbar()  # show color scale
```

```python
x = np.linspace(0, 10, 100)
dy = 0.1
y = np.sin(x) + dy * np.random.randn(100)

plt.errorbar(x, y, yerr=dy, fmt='.b');
```
